```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset1/master/abalone.csv")
```

```python
df.Sex.unique()
```

```python
colors=['r','g','b']
df.groupby('Sex').size().plot.bar(title='Total number of abalone represented by sex',color=colors)
```

```python
corr_df = df.drop(columns=['Sex'])

plt.figure(figsize=(19, 10))
corr = corr_df.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation heatmap (infants included)')
plt.show()
```

```python
corr_df = df.loc[df.Sex!='I'].drop(columns=['Sex'])

plt.figure(figsize=(19, 10))
corr = corr_df.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation heatmap (infants excluded)')
plt.show()
```

```python
def graph_sex(col):   
    fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(15,4))
    
    lower = df[col].min()
    upper = df[col].max()
    
    df.loc[df.Sex=='M'][col].plot.hist(range=[lower, upper], 
                                       grid=True, 
                                       color='b', 
                                       alpha=.5, 
                                       ax=ax1, 
                                       title=f'{col} frequencies by sex')
    
    df.loc[df.Sex=='F'][col].plot.hist(range=[lower, upper], 
                                       grid=True, 
                                       color='r', 
                                       alpha=.5, 
                                       ax=ax1)
    
    df.loc[df.Sex!='I'].groupby('Sex')[col].mean().plot.bar(ax=ax2, 
                                                            color=['r','b'], 
                                                            title=f'Avg. {col} by sex')
    
    df.loc[df.Sex!='I'].groupby('Sex')[col].median().plot.bar(ax=ax3, 
                                                              color=['r','b'], 
                                                              title=f'Median {col} by sex')
    
    plt.tight_layout()
graph_sex('Length')
```

```python
graph_sex('Diameter')
```

```python
graph_sex('Height')
```

```python
graph_sex('Whole weight')
```

```python
graph_sex('Shucked weight')
```

```python
graph_sex('Viscera weight')
```

```python
graph_sex('Shell weight')
```

```python
graph_sex('Rings')
```
